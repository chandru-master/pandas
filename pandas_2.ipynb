```python
import pandas as pd
```

```python
#now we gonna merge the table:
#dictionary method
df1 = pd.DataFrame ({
    "city":["chennai","madurai","kovai","tricy"],
    "temperature":[32,36,31,30],
})
df1
```

```python
df1.columns
```

```python
df2 = pd.DataFrame ({
    "city":["kovai","madurai","chennai","thanjur"],
    "humidity": [61,65,72,64],
})
df2
```

```python
#the output shows only which is common from both the table 
# data_name = pd.merge(df1,df2,on=)
df3 = pd.merge(df1,df2,on="city")
df3
```

```python
#the output shows only which is common from both the table 
df3 = pd.merge(df1,df2,on="city", how="inner")
df3
```

```python
#the output shows all which is  from both the table what exactly it has 
df3 = pd.merge(df1,df2,on="city", how="outer")
df3
```

```python
#the output will came through the order which we merge df1,df2 so df1 will be the result
df3 = pd.merge(df1,df2,on="city", how="left")
df3
```

```python
#the output will came through the order which we merge df1,df2 so df2 will be the result
df3 = pd.merge(df1,df2,on="city", how="right")
df3
```

```python
pd.merge()
```

```python
#indicators tell that how the output was taken from the table
df3 = pd.merge(df1,df2,on="city", how="outer",indicator=True)
df3
```

```python
df1 = pd.DataFrame ({
    "city":["chennai","madurai","kovai","tricy"],
    "temperature":[32,36,31,30],
    "sky":["clear","rainy  ","cloudy","dark"],
})
df1
```

```python
df2 = pd.DataFrame ({
    "city":["kovai","madurai","chennai","thanjur"],
    "humidity": [61,65,72,64],
    "sky":["cloudy","rainy","clear","dark"],
})
df2
```

```python
df3 = pd.merge(df1,df2,on="city")
df3
```

```python
df3 = pd.merge(df1,df2,on="city",suffixes=("_left","_right"))
df3
```
